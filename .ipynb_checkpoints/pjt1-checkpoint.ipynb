```python
# function for load the data 

def load_data(file_name):
    with open(file_name,"r") as file:
        data = json.load(file)
    return data
```

```python
# make a function which display the users friends and pages

def about_user(data):
   print("User and his connections\n")
   for user in data['users']:
       print(f"User ID : {user['id']} - {user['name']} is friends of : {user['friends']} and he likes the pages : {user['liked_pages']}")
   print("\nAbout page information:")
   for page in data['pages']:
        print(f"{page['id']} : {page['name']} ")
       
about_user(data)   # call the about_user function
```
